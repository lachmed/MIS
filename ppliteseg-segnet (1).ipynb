```python
import os
import numpy as np
import pandas as pd
import random

imagesNames = np.array(list(os.walk("/kaggle/working/PaddleSeg/data/images"))[0][2])

random.shuffle(imagesNames)
```

```python
listsPath = "/kaggle/working/PaddleSeg/data/"
imagesPath = "/kaggle/working/PaddleSeg/data/images/"
labelsPath = "/kaggle/working/PaddleSeg/data/labels/"

list_train = open(listsPath+"list_training.txt","w")
list_val = open(listsPath+"list_val.txt","w")
list_test = open(listsPath+"list_test.txt","w")

cpt = 0
for n in imagesNames:    
    cpt+=1
    if(cpt<=(70 * len(imagesNames))/100):        
        list_train.write("images/"+n+" "+"labels/"+n+"\n")    
    elif(cpt<=(90 * len(imagesNames))/100):        
        list_val.write("images/"+n+" "+"labels/"+n+"\n")
    else:
        list_test.write("images/"+n+" "+"labels/"+n+"\n")
                

```
